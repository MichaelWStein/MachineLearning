```python
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(20)
```

```python

#Creating the results
y = df["koi_disposition"]
print(y)
```

```python
#Creating the frame for x

df = df.drop(columns = ["koi_disposition"])

print(df.columns)

X = df [['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 'koi_period_err1', 'koi_period_err2',
       'koi_time0bk', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_impact',
       'koi_impact_err1', 'koi_impact_err2', 'koi_duration',
       'koi_duration_err1', 'koi_duration_err2', 'koi_depth', 'koi_depth_err1',
       'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2',
       'koi_teq', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2',
       'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_steff_err1',
       'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2',
       'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'ra', 'dec',
       'koi_kepmag']]

print(X.shape)
```

```python
#Preprocessing done before the Train Test Split 

import sklearn.preprocessing as skprep
scaler = skprep.MinMaxScaler()
print(scaler)
scaler.fit(X)
X = scaler.transform(X)
print(X)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
#Checking the shapes
print(X_train.shape, y_train.shape)
```

# Train the Support Vector Machine

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf.fit(X_train, y_train)
predictions = rf.predict(X_test)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, df.columns), reverse=True)
```

```python
print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchSearchCV` to tune the `n_estimators` and `max_features` parameters

```python
from pprint import pprint
# Look at parameters used by our current forest - n_estimators has been set to 200
print('Parameters currently in use:\n')
pprint(rf.get_params())
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [100, 200, 300, 500],
              'max_features': ['auto', 'sqrt']}
grid = GridSearchCV(rf, param_grid, verbose=3)
```

```python
# Train the model with GridSearch

# Fit the model using the grid search estimator. 
# This will take the SVC model and try each combination of parameters
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["CONFIRMED", "CANDIDATE", "FALSE POSITIVE"]))
```
